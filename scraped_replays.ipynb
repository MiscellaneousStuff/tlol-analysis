## Load JSON Replays

```python
import json
import os
import pandas as pd
json_replays_dir = "/Users/joe/Downloads/JSON Replays"
json_replays = os.listdir(json_replays_dir)
```

```python
len(json_replays)
```

## Load Dataframe

### Load Game Dataframe

```python
full_json_path = os.path.join(
    json_replays_dir,
    json_replays[0])
game_df = pd.read_json(full_json_path)
```

```python
game_df
```

### Load Champ Dataframe

```python
champs_lst = [ch for ch in game_df["champs"]]
flat_list = [item for sublist in champs_lst for item in sublist]
champs_df = pd.read_json(json.dumps(flat_list))
champs_df
```

```python
champs_df.columns
```

```python
ezreal_df = champs_df[champs_df["name"] == "Ezreal"]
ezreal_df
```

### Analyse Ezreal Dataframe

```python
ezreal_x = ezreal_df[ezreal_df["pos_x"] < 16000.0]["pos_x"]
ezreal_y = ezreal_df[ezreal_df["pos_z"] < 16000.0]["pos_z"]
```

```python
import matplotlib.pyplot as plt
plt.scatter(ezreal_x, ezreal_y, s=0.5)
```

### Load Minions Dataframe

```python
minions_lst = [ch for ch in game_df["minions"]]
flat_list = [item for sublist in minions_lst for item in sublist]
minions_df = pd.read_json(json.dumps(flat_list))
minions_df
```
